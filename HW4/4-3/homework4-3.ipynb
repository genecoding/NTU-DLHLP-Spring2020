## Q1

Please give some examples predicted correctly and incorrectly respectively. At least one for each case is required. Screenshot recommended.

```python
import json
import csv

with open('/content/hw4-3_test.json') as f:  
  d_test = json.load(f)
```

```python
print('Example of predicted correctly')  # 'id': '1151-10'
print()

print('Question')
for q in d_test['data'][0]['paragraphs'][0]['qas']:
  print(f"{q['id']}: {q['question']}")

print('Answer')
with open('predictions.csv') as f:
  pred = csv.DictReader(f)
  for p in pred:
    if '1151-10' in p['ID']:
      print(f"{p['ID']}: {p['Answer']}")

print()
print('Context')
d_test['data'][0]['paragraphs'][0]['context']
```

```python
print('Example of predicted incorrectly')  # 'id': '1151-13'
print()

print('Question')
for q in d_test['data'][0]['paragraphs'][2]['qas']:
  print(f"{q['id']}: {q['question']}")

print('Answer')
with open('predictions.csv') as f:
  pred = csv.DictReader(f)
  for p in pred:
    if '1151-13' in p['ID']:
      print(f"{p['ID']}: {p['Answer']}")

print()
print('Context')
d_test['data'][0]['paragraphs'][2]['context']
```
