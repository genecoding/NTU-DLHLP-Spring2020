## Q4
思考一下為何有無PIT會影響3-1, 3-2的結果並寫下你的看法

In 3-1, since there are only 2 speakers and the dataset is speaker-dependent, the performance of training without PIT is still fine, and the performance of training with PIT is even better. When it comes to 3-2, since there are multiple speakers and the dataset is speaker-independent, the performance of training without PIT declines incredibly, which shows PIT is important for speech separation.
